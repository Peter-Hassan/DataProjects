### API Call to DataFrame function

```python
def runelite_price_data(itemid):
    headers = {
    'User-Agent': 'My User Agent 1.0',
    'From': 'youremail@domain.example'  # This is another valid field
    }
    global item_ids

    r = requests.get(f'https://prices.runescape.wiki/api/v1/osrs/timeseries?timestep=24h&id={itemid}', headers=headers)
    data = json.loads(r.text)
    runelite = pd.DataFrame(data)
    flattened_data = []
    for item in data['data']:
        flattened_data.append(item)
    df = pd.DataFrame(flattened_data)
    df['timestamp'] = pd.to_datetime(df['timestamp'], unit='s')
    df['avgHighPrice'] = df['avgHighPrice'].fillna(method='ffill')
    df['avgLowPrice'] = df['avgLowPrice'].fillna(method='ffill')
    columns_to_drop = ['highPriceVolume', 'lowPriceVolume']
    df.drop(columns_to_drop, axis=1, inplace=True)
    return df
```

### Creating Merged Dataframe

```python
def get_item_name(item):
    name = f2p[f2p['Item ID'] == item]['Item Name'].values[0]
    return name

merged_df = pd.DataFrame()
number = 0
dfs = []
keys = []

for index, item in enumerate(item_ids):
    try:
        # Data
        x = runelite_price_data(item_ids[index])
        dfs.append(x)
        # Keys
        name = get_item_name(item)
        keys.append(name)

        # Position
        print(f'{item} : {index+1}/{len(item_ids)}')   

        # Display
        number += 1
        if number == 5: 
            clear_output()
            number = 0
    except Exception as e:
        print(f'Item not appended: {item}, {e}')
clear_output()
        
        
merged = pd.concat(dfs, axis=1, keys = keys)

# Set column names for MultiIndex dataframe
merged.columns = merged.columns.set_names(['Name', 'Data'])

# Format the index to be the timestamp
merged.set_index((merged.columns[0][0], 'timestamp'), inplace=True)

# Rename Index name
merged.index.name = 'Date'

# Format as datetime
merged.index = pd.to_datetime(merged.index)
merged.head(5)
```

```python
# Looking at only average high price data
refined = merged.xs(key='avgHighPrice', axis=1, level='Data').copy()
refined.head(5)
```

```python
fig = plt.figure(figsize=(10,10), dpi=100)

sns.heatmap(refined.corr(), cmap='coolwarm', annot=True)
```

```python
# Assuming refined is the DataFrame containing the data series: 'Mithril ore' and 'Silver ore'
x = refined.index  # Assuming the index of the DataFrame is the shared x-axis values
y1 = refined['Mithril ore']
y2 = refined['Silver ore']

fig, ax1 = plt.subplots()

ax1.plot(x, y1, color='blue', label='Mithril ore')
ax1.set_xlabel('X-axis label')
ax1.set_ylabel('Mithril Ore', color='blue')

ax2 = ax1.twinx()
ax2.plot(x, y2, color='red', label='Silver ore')
ax2.set_ylabel('Silver Ore', color='red')

# Additional customization
plt.title('Mithril and Silver Ore Comparison')
plt.legend()

# Display the plot
plt.show()
```

### Estimate for profits selling items to store

```python
mithril_ore = 100
onyx = 1790000
tokkul_mith = 16
tokkul_onyx = 260000

tokkul_cost_per_mith = mithril_ore/tokkul_mith
tokkul_profit_per_onyx = onyx/tokkul_onyx
profit_per_tokkul = tokkul_profit_per_onyx-tokkul_cost_per_mith
print(f'Profit for 1 overstocked tokkul {profit_per_tokkul}')

tics_hour = 0.6*60*60
sells_a_tic = 50*3
max_sales = tics_hour*sells_a_tic
profit = max_sales * 16 * profit_per_tokkul
print(f'Max Sales made in an hour: {max_sales} \n')
profit
```
